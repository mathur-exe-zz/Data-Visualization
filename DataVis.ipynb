```python
# import torch, torchvision, torchaudio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yeild (tons per hectare)')
plt.legend(['Apples', 'Oranges'])
plt.title('Crop Yield');
```

```python
flower_df = sns.load_dataset('iris')
flower_df
```

```python
species = flower_df['species'].unique()
species
```

```python
sns.set_style('darkgrid')
```

```python
plt.plot(flower_df['sepal_length'], flower_df['sepal_width']);
```

```python
sns.scatterplot(flower_df['sepal_length'], flower_df['sepal_width']);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimnesion')
sns.scatterplot(flower_df['sepal_length'], 
                flower_df['sepal_width'], 
                hue=flower_df['species'], 
                s=140);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimnesion')
sns.scatterplot('sepal_length', 
                'sepal_width', 
                hue=flower_df['species'], 
                s=140, 
                data=flower_df);
```

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
virginica_df = flower_df[flower_df.species == 'virginica']
```

```python
plt.hist([setosa_df['sepal_width'], versicolor_df['sepal_width'], virginica_df['sepal_width']],
        bins=np.arange(2, 4.5, 0.25), 
        stacked=True)
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
# plt.figure(figsize=(18, 12));
```

```python
plt.hist(setosa_df['sepal_width'], alpha=0.4, 
        bins=np.arange(2, 4.5, 0.25))
plt.hist(versicolor_df['sepal_width'], alpha=0.4, bins=np.arange(2, 4.5, 0.25))
# plt.hist(virginica_df['sepal_width'], alpha=0.4, bins=np.arange(2, 4.5, 0.25))
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
plt.figure(figsize=(18, 12));
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
avg_bill_df = tips_df.groupby('day')[['total_bill']].mean()
avg_bill_df
```

```python
plt.bar(avg_bill_df.index, avg_bill_df['total_bill'])
```

```python
sns.barplot('day', 'total_bill', data=tips_df);
```
